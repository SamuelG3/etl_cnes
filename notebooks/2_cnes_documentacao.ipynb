```python
import docx
import pandas as pd
import os
import json
import numpy as np
```

```python
folder_docs = r"../documentos"
```

```python
# Define o caminho para o arquivo docx de dicionário de dados
doc_path = os.path.join(folder_docs, 'FTP-CNES-Files', "Arquivos extraídos", "DICIONARIO_DE_DADOS.docx")

# Extrai todas as tabelas do documentos
def extract_all_tables(doc_path):
    document = docx.Document(doc_path)
    all_tables = []
    for table in document.tables:
        table_data = [[cell.text for cell in row.cells] for row in table.rows]
        all_tables.append(table_data)
    return all_tables

tbls_docx = []
all_tables_data = extract_all_tables(doc_path)

# printa e adiciona a tabela em uma nova lista
for idx, table in enumerate(all_tables_data):
    print(f"Table {idx + 1}:")
    print(table)
    print("\n---\n")
    tbls_docx.append(table)
```

```python
# Corrige as tabelas que foram lidas de forma fora do padrão adicionando ou removendo valores
tbls_docx.insert(0, 'RELAÇÃO DAS TABELAS PRINCIPAIS')
tbls_docx.insert(6, 'LFCES002 - RL_ESTAB_COMPLEMENTAR')
tbls_docx.pop(83)
tbls_docx.insert(156, 'RELAÇÃO DAS TABELAS DE DOMÍNIOS')
tbls_docx.pop(172)
tbls_docx.insert(184, 'NFCES015 - TB_SERVICO_APOIO')
tbls_docx.insert(186, 'NFCES017 - TB_COMISSAO_OUTRO')
tbls_docx.insert(188, 'NFCES018 - TB_CLASSIFICACAO_SERVICO')
tbls_docx.insert(190, 'NFCES019 - TB_SERVICO_ESPECIALIZADO')
tbls_docx.insert(192, 'NFCES020 - TB_SUBTIPO_INSTALACAO')
tbls_docx.insert(194, 'NFCES021 - TB_INSTAL_FISICA_PARA_ASSIST')
tbls_docx.insert(196, 'NFCES022 - TB_TIPO_SERVICO_RFERENCIADO')
```

```python
# Função para corrigir tabelas que tenham apenas o código e nome da tabela
def flatten_and_concatenate(nested_list):

    flattened_list = [item for sublist in nested_list for item in sublist if isinstance(sublist, list)]
    concatenated_string = ' - '.join(str(item) for item in flattened_list)
    
    return concatenated_string

cleaned_list = []

# Para cada tabela par utiliza a função flatten_and_concatenate para transformar a tabela em uma string e adiciona os valores na lista limpa
for index, item in enumerate(tbls_docx):
    if index % 2 == 0:
        if isinstance(item, list):
            try:
                result_string = flatten_and_concatenate(item)
                cleaned_list.append(result_string)
            except Exception as e:
                print(item)
                print(e)
        else:
            cleaned_list.append(item)
    else:
        cleaned_list.append(item)
```

```python
result_dict = {}  # Cria uma nova dicionário vazio chamado result_dict

for i in range(0, len(cleaned_list), 2):  # Inicia um loop usando range com passo 2
    key = cleaned_list[i]  # Define a chave como o elemento na posição atual
    value = cleaned_list[i+1]  # Define o valor como o elemento seguinte
    result_dict[key] = value  # Adiciona a chave e o valor ao dicionário result_dict

for key, value in result_dict.items():  # Inicia outro loop para iterar sobre os itens do dicionário
    print(key)  # Imprime a chave
    print("\n---\n")  # Imprime uma linha de separação
```

```python
for key, value in result_dict.items():
    try:
        temp_df = pd.DataFrame(value)
        temp_df.to_csv(f'{folder_docs}/documentacao_cnes/{key.strip()}.csv', sep=';', index=False)
    except Exception as e:
        print(f'Erro em: {e}')
        print(value)
        print('\n --- \n')
```

```python
# Lê a tabela com a documentação de todas as tabelas para padronizar os arquivos
df_tabelas = pd.DataFrame()
arquivos_tabelas = {
    'DOMINIO': "RELAÇÃO DAS TABELAS DE DOMÍNIOS",
    'PRINCIPAL': "RELAÇÃO DAS TABELAS PRINCIPAIS"
}

for key, value in arquivos_tabelas.items():    
    temp_df = pd.read_csv(f'{folder_docs}/documentacao_cnes/{value}.csv', sep=';', skiprows=1, dtype=str)
    temp_df.insert(0, 'TIPO DE TABELA', key)
    df_tabelas = pd.concat([df_tabelas, temp_df])

df_tabelas.sample(5)
```

```python
# Corrige nomes para keys e converte o DataFrame para um dicionário
df_tabelas.rename(columns={
    "TIPO DE TABELA": "tipo_tabelas", 
    "NOME DA TABELA (BANCO LOCAL)": "nome_tabela_banco_local", 
    "NOME DA TABELA (BANCO DE PRODUÇÃO FEDERAL)": "nome_tabela_banco_prod_fed",
    "DESCRIÇÃO DA TABELA": "desc_tabela"
    }, inplace=True)

for col in df_tabelas:
    df_tabelas[col] = df_tabelas[col].str.strip()

dict_tabelas = df_tabelas.to_dict(orient='records')

# Salvar o dicionário como JSON
try:
    with open(f'{folder_docs}/documentacao_cnes/json/relacao_tabelas.json', 'w') as f:
        json.dump(dict_tabelas, f, indent=4)

    print("DataFrame salvo como JSON com sucesso!")
except Exception as e:
    print(f"Um erro ocorreu: {str(e)}")
```

```python
list_cod_nomes_tb = df_tabelas["nome_tabela_banco_local"].to_list()
list_cod_nomes_tb
```

```python
folder_path = folder_docs + "/documentacao_cnes"

for file in os.listdir(folder_path):
    for name in list_cod_nomes_tb:
        if name.upper() in file.upper():
            old_file_path = os.path.join(folder_path, file)
            new_filename = f"{name}.csv"
            
            new_file_path = os.path.join(folder_path, new_filename)
            try:
                os.rename(old_file_path, new_file_path)
                print(f"Renomeado '{file}' para '{new_filename}'")
            except FileNotFoundError:
                print(f"Erro: Arquivo '{file}' não foi encontrado na pasta.")
            except PermissionError:
                print(f"Permissão negada: Não foi possível renomear '{file}'.")
            except Exception as e:
                print(f"Um erro ocorreu  '{file}': {str(e)}")

print("Arquivos renomeados com sucesso.")
```

```python
file_for_example = r"..\documentos\documentacao_cnes\NFCES021.csv"
def manipula_e_trata_csv(df):

    for col in df.columns:
        df[col] = df[col].str.strip()

    for col in df.columns[0:9]:
        df[col] = df[col].apply(lambda x: x.replace('\n', '') if isinstance(x, str) else x)

    mask = df[df['0'] == "NOME DO CAMPO"].index[0]
    df = df.loc[mask:].reset_index(drop=True)
    
    df.columns = df.iloc[0]

    if "DOMÍNIOS" in df.columns:
        df["DOMÍNIOS"] = df["DOMÍNIOS"].str.replace("\n", ",")

    df = df.dropna(axis=1, how='all')
    df = df.iloc[1:]
    df = df.rename(columns={np.nan: 'IS NULL'})
    if "IS NULL" not in df.columns:
        df.insert(7, "IS NULL", np.nan)

    return df

# Uso de exmplo
df_example = pd.read_csv(file_for_example, sep=";", dtype=str)
df_example = manipula_e_trata_csv(df_example)
df_example.head(5)
```

```python
folder_path = folder_docs + "/documentacao_cnes"

exceptions = [
    "RELAÇÃO DAS TABELAS DE DOMÍNIOS.csv",
    "RELAÇÃO DAS TABELAS PRINCIPAIS.csv"
    ]

for file in os.listdir(folder_path): 
    if file not in exceptions and file.endswith(".csv"):
        try:
            temp_df = pd.read_csv(folder_path + "/" + file, sep=";", dtype=str)
            temp_df = manipula_e_trata_csv(temp_df)
            temp_df.to_csv(folder_path + "/" + file, sep=";", index=False)
        except FileNotFoundError:
            print(f"Erro: Arquivo '{file}' não encontrado no caminho.")
        except PermissionError:
            print(f"Permissão negada: Não foi possível abrir o arquivo '{file}'.")
        except Exception as e:
            print(f"Um erro ocorreu enquanto operava o arquivo '{file}': {str(e)}")
```
