```python
import datetime
import pandas as pd
from ftplib import FTP
import zipfile
import os
```

```python
### Define o caminho da pasta
folder_docs = r"..\documentos"
```

```python
### Define o datetime
today = datetime.date.today()
first = today.replace(day=1)
last_month = first - datetime.timedelta(days=1)
month = last_month.strftime('%m')
```

```python
### Define os arquivos para o ftp
files_ftp = [
    'SCNES_DICIONARIO_DE_DADOS.ZIP',
    f'BASE_DE_DADOS_CNES_{last_month.year}{month}.zip'
]

# Define caminho para o FTP
ftp = FTP('ftp.datasus.gov.br')
login_result = ftp.login()

for f in files_ftp:
    local_file_path = os.path.join(folder_docs, 'FTP-CNES-Files', f)

    if 'BASE_DE_DADOS_CNES_' in f:
        ftp.cwd('/cnes')
    else:
        ftp.cwd('/cnes/Download')

    try: 
        with open(local_file_path, 'wb') as fp:
            ftp.retrbinary(f'RETR {f}', fp.write)

        print(f"Arquivo {f} foi baixado com sucesso.")
    except Exception as e:
        print(f"{f} não pode ser baixado com sucesso. Erro: {e}")

ftp.quit()
```

```python
# Especifica os caminhos
zip_file_path = os.path.join(folder_docs, 'FTP-CNES-Files', f"BASE_DE_DADOS_CNES_{last_month.year}{month}.zip")
target_folder = os.path.join(folder_docs, 'FTP-CNES-Files', "Arquivos extraídos", f"BASE_DE_DADOS_CNES_{last_month.year}{month}")

# Cria o target folder caso ele não exista
if not os.path.exists(target_folder):
    os.makedirs(target_folder)

# Abre o arquivo zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    try:
        # Extraí todo o conteúdo para o caminho
        zip_ref.extractall(target_folder)
        print(f"Extraído conteúdo com sucesso para {target_folder}")
    except zipfile.BadZipFile:
        print(f"Erro: {zip_file_path} não é um arquivo zip válido")
    except PermissionError:
        print(f"Erro: Não possuí permissão para escrever no caminho {target_folder}")
    except Exception as e:
        print(f"Um erro inesperado ocorreu: {e}")
```
